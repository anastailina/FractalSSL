```python
#──────────────────────────────────────────────────────────────────────────
# 0.  Imports & path configuration
# ─────────────────────────────────────────────────────────────────────────────
from pathlib import Path
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader, random_split

BASE = PADS_PREPROCESSED   # adjust if different
MOVEMENT_DIR = BASE / "movement"
CSV = BASE / "file_list.csv"

assert MOVEMENT_DIR.exists() and CSV.exists(), "check your path ↑"


# ─────────────────────────────────────────────────────────────────────────────
# 1.  Inspect ONE file to understand the binary layout
#     (float32  →  n_channels × 1024 samples, 10.24 s @ 100 Hz)
# ─────────────────────────────────────────────────────────────────────────────
N_CHAN_MOVEMENT = 132              # fixed by dataset design

def load_bin(file: Path, n_chan: int = N_CHAN_MOVEMENT) -> np.ndarray:
    """
    Return an array shaped (channels, time).
    Works for all movement *.bin files even though time-length varies (≈ 9.8–10.2 s).
    """
    arr = np.fromfile(file, dtype=np.float32)
    if arr.size % n_chan:
        raise ValueError(f"{file.name}: {arr.size} values not divisible by {n_chan}")
    n_time = arr.size // n_chan
    return arr.reshape(n_chan, n_time)

# test
x = load_bin(Path("/content/drive/MyDrive/pads-parkinsons-disease-smartwatch-dataset-1.0.0/preprocessed/movement/001_ml.bin"))
print(x.shape)        # → (132, 976)



```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 2.  Minimal Dataset object
# ─────────────────────────────────────────────────────────────────────────────
class PADS(Dataset):
    """
    Each sample = (T, C) float32 tensor + label + pid
    Shape is transposed to (time, channels) because TS2Vec & friends
    expect that ordering by default.
    """
    def __init__(self, root: Path, ids: list[int], transform=None):
        self.root, self.transform = root, transform
        df = pd.read_csv(root / "file_list.csv")
        self.df = df[df["id"].isin(ids)].reset_index(drop=True)

        self.paths  = [root / "movement" / f"{pid:03d}_ml.bin" for pid in self.df["id"]]
        self.labels = self.df["label"].astype(int).tolist()
        self.pids   = self.df["id"].astype(int).tolist()

    def __len__(self): return len(self.paths)

    def __getitem__(self, idx):
        x = load_bin(self.paths[idx])                      # (C, T)
        if self.transform: x = self.transform(x)
        x = torch.from_numpy(x.T)                          # → (T, C)
        return x, self.labels[idx], self.pids[idx]
```

```python
BATCH_SIZE = 128

# ╭──────────────────────────────────────────────────────────────────────────╮
# │ 3.  Leave-one-subject-out split utility                                  │
# ╰──────────────────────────────────────────────────────────────────────────╯
def make_loaders(leave_out_pid: int):
    all_ids = pd.read_csv(BASE / "file_list.csv")["id"].astype(int).tolist()
    train_ids = [pid for pid in all_ids if pid != leave_out_pid]
    val_ids   = [leave_out_pid]

    train_ds = PADS(BASE, train_ids)
    val_ds   = PADS(BASE, val_ids)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE,
                              shuffle=True,  num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE,
                              shuffle=False, num_workers=2, pin_memory=True)
    return train_loader, val_loader

# ╭──────────────────────────────────────────────────────────────────────────╮
# │ 4.  Example: pick one held-out patient and feed to your SSL pipeline     │
# ╰──────────────────────────────────────────────────────────────────────────╯
LOSO_PID = 7                                   # try 1…492
train_loader, val_loader = make_loaders(LOSO_PID)

# sanity-check a single batch
xb, yb, pids = next(iter(train_loader))
print("batch:", xb.shape, yb[:5], pids[:5])    # (B, T, C)  eg  (128, 976, 132)
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 0.  Imports & paths
# ─────────────────────────────────────────────────────────────────────────────
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit
import torch
from ts2vec import TS2Vec
# import categorical
from torch.utils.data import Dataset, DataLoader
         #


BASE = PADS_PREPROCESSED   # adjust if different
CSV = BASE / "file_list.csv"  # adjust if needed
BIN_DIR = BASE / "movement"

N_CHAN       = 132         # fixed: 11 tasks × 2 wrists × 2 sensors × 3 axes
TARGET_LEN   = 1024        # pad / crop every sample to this length
RANDOM_SEED  = 42          # keep experiments reproducible
TEST_SIZE    = 0.2

# ─────────────────────────────────────────────────────────────────────────────
# 1.  Helper: load & length-normalise one .bin file
# ─────────────────────────────────────────────────────────────────────────────
def load_bin(path: Path, n_chan: int = N_CHAN, tgt_len: int = TARGET_LEN) -> np.ndarray:
    buf = np.fromfile(path, dtype=np.float32)
    if buf.size % n_chan:
        raise ValueError(f"{path.name}: {buf.size} values not divisible by {n_chan}")
    n_time = buf.size // n_chan
    x = buf.reshape(n_chan, n_time)           # (C, T_now)

    if n_time < tgt_len:                      # zero-pad short recordings
        pad = np.zeros((n_chan, tgt_len), dtype=np.float32)
        pad[:, :n_time] = x
        x = pad
    elif n_time > tgt_len:                    # crop long recordings
        x = x[:, :tgt_len]

    return x.T                                # (T, C)  → TS2Vec default

# ─────────────────────────────────────────────────────────────────────────────
# 2.  Load *all* samples into one (N, T, C) array  (≈ 500 × 1024 × 132, ~260 MB)
# ─────────────────────────────────────────────────────────────────────────────
meta = pd.read_csv(BASE / "file_list.csv")    # columns: id, label, ...
paths   = [BIN_DIR / f"{pid:03d}_ml.bin" for pid in meta["id"]]
meta["condition_label"] = meta["condition"].astype("category").cat.codes
labels  = meta["label"].astype(int).values    # 0 HC, 1 PD, 2 DD
#labels = meta["condition_label"].values
groups  = meta["id"].astype(int).values       # patient id for LOSO / grouping

windows = np.stack([load_bin(p) for p in paths]).astype(np.float32)
# shape → (n_samples , 1024 , 132)

```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 3.  Stratified *group* split  (same function you already use)
# ─────────────────────────────────────────────────────────────────────────────
def stratified_group_split(X, y, groups, test_size=0.2, seed=42, max_trials=1_000):
    rs = np.random.RandomState(seed)
    all_classes = np.unique(y)
    for _ in range(max_trials):
        random_state = rs.randint(0, 10_000_000)
        gss = GroupShuffleSplit(n_splits=1, test_size=test_size,
                                random_state=random_state)
        tr_idx, te_idx = next(gss.split(X, y, groups=groups))
        if np.all(np.isin(all_classes, y[tr_idx])):
            return tr_idx, te_idx
    raise RuntimeError("Could not find a split with all classes in train")

tr_idx, te_idx = stratified_group_split(windows, labels, groups,
                                        test_size=TEST_SIZE, seed=RANDOM_SEED)

# (optional) keep the indices for later
np.save('/content/drive/MyDrive/FractalSSL_backup/models/pads_train_idx.npy', tr_idx)
np.save('/content/drive/MyDrive/FractalSSL_backup/models/pads_test_idx.npy',  te_idx)

# ─────────────────────────────────────────────────────────────────────────────
# 4.  Train TS2Vec exactly the way you did for PAMAP2
# ─────────────────────────────────────────────────────────────────────────────
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

enc = TS2Vec(
    input_dims=N_CHAN,     # 132
    output_dims=320,
    device=device,
    batch_size=64,
    lr=1e-3
)

enc.fit(windows[tr_idx], n_epochs=40, verbose=True)
enc.save('/content/drive/MyDrive/FractalSSL_backup/models/pads_ts2vec_backbone.pt')
```

## Fractal SSL Fine-Tuning

```python
def nt_xent(reps: torch.Tensor, batch_size: int, n_views: int, T: float) -> torch.Tensor:
    """
    reps      : (B * n_views, D) ℓ2-normalised
    returns   : scalar loss
    """
    N   = reps.size(0)
    sim = reps @ reps.T / T                       # cosine sim over temperature

    # zero out self-similarities (i == j)
    sim.fill_diagonal_(-1e4)

    # positive index: for each view i, its twin is (i ± 1) within the same sample
    view_idx   = torch.arange(N, device=reps.device)
    pos_index  = view_idx ^ 1                     # flips last bit: 0↔1, 2↔3, …
    loss       = torch.nn.functional.cross_entropy(sim, pos_index)

    return loss
```

```python
import torch, random, numpy as np
from typing import List, Tuple

class FractalViewGenerator:
    def __init__(self, levels: int = 4):
        self.levels = levels
    def __call__(self, ts: torch.Tensor) -> Tuple[List[torch.Tensor], List[int]]:
        L, views, lens = ts.size(0), [], []
        for i in range(self.levels):
            win   = int(L / 2 ** i)
            start = random.randint(0, L - win)
            crop  = ts[start:start + win]
            views.append(crop)
            lens.append(win)
        return views, lens

class FractalSSL(torch.nn.Module):
    def __init__(self, backbone, proj_dim=128, levels=4,
                 temperature=0.1, freeze_backbone=True):
        super().__init__()
        self.backbone, self.levels, self.temperature = backbone, levels, temperature
        self.view_gen = FractalViewGenerator(levels)

        # ---- determine channels & device from first param -------------------
        first_w  = next(backbone.net.parameters()) if hasattr(backbone, "net") \
                   else next(backbone.parameters())
        self._in_channels = first_w.shape[1] if first_w.ndim >= 2 else first_w.shape[0]

        # ---- optional freezing ---------------------------------------------
        if freeze_backbone and hasattr(backbone, "net"):
            for p in backbone.net.parameters():
                p.requires_grad = False

        # ---- infer latent dimension ----------------------------------------
        with torch.no_grad():
            dummy = np.zeros((1, 8, self._in_channels), dtype='float32')
            latent_dim = backbone.encode(dummy).shape[-1]

        self.proj = torch.nn.Sequential(
            torch.nn.Linear(latent_dim, proj_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(proj_dim, proj_dim)
        )

    # ---------- tiny helper so we never forget the NumPy conversion ----------
    def _ts2vec_encode(self, x: torch.Tensor) -> torch.Tensor:
      """
      Wrap TS2Vec.encode(): accepts (N,L,C) tensor, returns tensor on same device
      no matter whether the fork outputs NumPy or PyTorch.
      """
      dev   = x.device
      out   = self.backbone.encode(x.detach().cpu().numpy())  # NumPy in → ??? out
      feats = torch.as_tensor(out, device=dev, dtype=torch.float32)  # <── FIX
      return feats

    # ------------------------------------------------------------------------
    def forward(self, ts_batch: torch.Tensor) -> torch.Tensor:
        device, views, lens = ts_batch.device, [], []
        for ts in ts_batch:
            sub, ln = self.view_gen(ts)
            views += [v.to(device) for v in sub]
            lens  += ln

        padded = torch.nn.utils.rnn.pad_sequence(views, batch_first=True)
        mask   = (torch.arange(padded.size(1), device=device)[None, :]
                 < torch.tensor(lens, device=device)[:, None])

        emb = self._ts2vec_encode(padded)          # (N,T,D) on same device
        emb = (emb * mask.unsqueeze(-1)).sum(1) / mask.sum(1, keepdim=True)

        proj = torch.nn.functional.normalize(self.proj(emb), dim=-1)
        return proj

```

```python
import torch
import random
import numpy as np
from typing import List, Tuple

# ── 1. Nested (fractal) view generator ───────────────────────────────────────
class NestedViewGenerator:
    def __init__(self, levels: int = 4):
        self.levels = levels

    def __call__(self, ts: torch.Tensor) -> Tuple[List[torch.Tensor], List[int]]:
        """
        Produce `levels` nested crops:
         - view 0: full length [0 : L]
         - view 1: random start, length L/2
         - view 2: random start, length L/4
         - …
        Returns:
            views: list of T_i × C tensors
            lens:  their lengths
        """
        L = ts.size(0)
        views, lens = [], []
        # view 0 is always the full sequence
        views.append(ts)
        lens.append(L)
        # subsequent views get shorter, nested within the full sequence
        for i in range(1, self.levels):
            win = L // (2 ** i)
            start = random.randint(0, L - win)
            crop = ts[start : start + win]
            views.append(crop)
            lens.append(win)
        return views, lens


# ── 2. Multi‐positive NT‐Xent loss ─────────────────────────────────────────────
def multi_nt_xent(reps: torch.Tensor, B: int, levels: int, T: float) -> torch.Tensor:
    """
    reps: (B * levels, D) ℓ2‐normalized
    We reshape to (B, levels, D) and treat all level‐pairs as positives.
    """
    device = reps.device
    D = reps.size(-1)
    reps = reps.view(B, levels, D)           # → (B, levels, D)
    reps = torch.nn.functional.normalize(reps, dim=-1)

    # flatten back for easy sim computation
    flat = reps.view(B * levels, D)          # (N, D) with N = B*levels
    sim = flat @ flat.T / T                  # (N, N)

    # mask out self‐sims
    diag = torch.eye(B * levels, device=device) * 1e4
    sim = sim - diag

    # build targets: for each i in [0..B*levels), positives = {i’s other levels of same sample}
    labels = torch.arange(B, device=device).repeat_interleave(levels)
    # We’ll do a multi‐way cross‐entropy: for each anchor i, positives are j where labels[j] == labels[i]
    # Create mask:
    pos_mask = labels.unsqueeze(0) == labels.unsqueeze(1)   # (N, N)
    neg_mask = ~pos_mask

    # For stability, we can compute per‐anchor:
    logits = sim
    exp = torch.exp(logits)
    pos_exp = exp * pos_mask.float()
    neg_exp = exp * neg_mask.float()

    # sum over positives and negatives
    sum_pos = pos_exp.sum(1)
    sum_all = exp.sum(1)
    loss = -torch.log(sum_pos / sum_all).mean()
    return loss


# ── 3. FractalSSL with direct PyTorch backbone calls ──────────────────────────
class FractalSSL(torch.nn.Module):
    def __init__(self,
                 backbone_net: torch.nn.Module,
                 proj_dim: int = 128,
                 levels: int = 4,
                 temperature: float = 0.1,
                 freeze_backbone: bool = False):
        """
        backbone_net: the TS2Vec network itself (not encode()), expects (B, C, L) in / (B, L, C)
        """
        super().__init__()
        self.backbone = backbone_net
        self.levels = levels
        self.temperature = temperature
        self.view_gen = NestedViewGenerator(levels)

        # optional freezing of some layers
        if freeze_backbone:
            for p in self.backbone.parameters():
                p.requires_grad = False

        # infer latent dimension by a dummy forward
        with torch.no_grad():
            # suppose backbone takes (B, C, L) and returns (B, L, D)
            dummy = torch.zeros(1, backbone_net.net.in_channels, 8)  # adjust if needed
            latent = self.backbone.net(dummy)       # (1, D, 8) or (1, 8, D)
            if latent.ndim == 3 and latent.shape[1] == dummy.shape[1]:
                latent = latent.permute(0, 2, 1)
            D = latent.shape[-1]

        # simple two‐layer projector
        self.proj = torch.nn.Sequential(
            torch.nn.Linear(D, proj_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(proj_dim, proj_dim),
        )

    def forward(self, ts_batch: torch.Tensor) -> torch.Tensor:
        """
        ts_batch: (B, L, C)
        returns:  (B * levels, proj_dim) normalized projections
        """
        device = ts_batch.device
        views, lens = [], []

        # generate nested views for each sample
        for ts in ts_batch:
            vs, ls = self.view_gen(ts)  # list of (T_i, C)
            views += vs
            lens += ls

        # pad to the max time among all views
        padded = torch.nn.utils.rnn.pad_sequence(views, batch_first=True)  # (B*levels, T_max, C)
        # permute to (B*levels, C, T_max) if backbone expects channels-first:
        padded = padded.permute(0, 2, 1)

        # run backbone directly
        out = self.backbone.net(padded)       # e.g. (N, D, T_max)
        if out.ndim == 3:
            out = out.permute(0, 2, 1)        # → (N, T_max, D)

        # mask & mean‐pool each view to get (N, D)
        mask = torch.arange(out.size(1), device=device)[None, :] < torch.tensor(lens, device=device)[:, None]
        emb = (out * mask.unsqueeze(-1).float()).sum(1) / mask.sum(1, keepdim=True)

        # project & normalize
        proj = self.proj(emb)
        proj = torch.nn.functional.normalize(proj, dim=-1)
        return proj
```
